```python
# Perform basic arithmetic based on operator input

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == '+':
    print(f"{num1} + {num2} = {num1 + num2}")
elif op == '-':
    print(f"{num1} - {num2} = {num1 - num2}")
elif op == '*':
    print(f"{num1} * {num2} = {num1 * num2}")
elif op == '/':
    if num2 != 0:
        print(f"{num1} / {num2} = {num1 / num2}")
    else:
        print("Error: Division by zero!")
else:
    print("Invalid operator")
```

```python
# Check if triangle is equilateral, isosceles, or scalene

a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))

# First check if it forms a valid triangle
if a + b > c and a + c > b and b + c > a:
    
    if a == b == c:
        print("The triangle is Equilateral.")
    elif a == b or b == c or a == c:
        print("The triangle is Isosceles.")
    else:
        print("The triangle is Scalene.")

else:
    print("The given sides do not form a valid triangle.")
```

```python
# Calculate bill: first 100 units free, next 100 @5/unit, above @10/unit
units = int(input("Enter units consumed: "))

if units <= 100:
    bill = 0
elif units <= 200:
    bill = (units - 100) * 5
else:
    bill = (100 * 5) + (units - 200) * 10

print(f"Total Electricity Bill: Rs. {bill}")
```

```python
# Check if withdrawal is possible (balance check, minimum balance 500)

balance = 5000

try:
    withdrawal = int(input("Enter amount to withdraw: "))

    if withdrawal <= 0:
        print("Invalid amount")
    elif withdrawal > balance - 500:
        print("Insufficient balance. Minimum balance of 500 required")
    else:
        balance -= withdrawal
        print(f"Withdrawal successful! Remaining balance: {balance}")

except ValueError:
    print("Please enter a valid number.")
```

```python
# Calculate BMI and category (Underweight, Normal, Overweight, Obese)

weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height ** 2)
print(f"Your BMI: {round(bmi, 2)}")

if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal weight")
elif bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
```

```python
# Reverse a given number using while loop

num = int(input("Enter a number: "))
original = num
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10

print(f"Reverse of {original} is {reversed_num}")
```

```python
# Check if number is Armstrong (153 = 1³ + 5³ + 3³)

num = int(input("Enter a number: "))
original = num
sum_of_powers = 0

for digit in str(num):
    sum_of_powers += int(digit) ** len(str(num))

if sum_of_powers == original:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
# Print message based on HTTP status code

code = int(input("Enter HTTP status code: "))

match code:
    case 200:
        print("OK - Request successful")
    case 201:
        print("Created - Resource created successfully")
    case 301:
        print("Moved Permanently - Resource has moved")
    case 302:
        print("Found - Resource found temporarily")
    case 400:
        print("Bad Request - Server couldn't understand request")
    case 401:
        print("Unauthorized - Authentication required")
    case 403:
        print("Forbidden - Access denied")
    case 404:
        print("Not Found - Resource doesn't exist")
    case 500:
        print("Internal Server Error - Server error")
    case 502:
        print("Bad Gateway - Invalid response from upstream")
    case 503:
        print("Service Unavailable - Server is down")
    case _:
        print(f"{code} - Unknown status code")
```

```python
# Calculate area based on shape input

shape = input("Enter shape (circle, square, rectangle): ").lower()

match shape:
    case "circle":
        radius = float(input("Enter radius: "))
        area = 3.14159 * radius ** 2
        print(f"Area of circle: {area:.2f}")
    
    case "square":
        side = float(input("Enter side length: "))
        area = side ** 2
        print(f"Area of square: {area}")
    
    case "rectangle":
        length = float(input("Enter length: "))
        width = float(input("Enter width: "))
        area = length * width
        print(f"Area of rectangle: {area}")
    
    case _:
        print("Invalid shape")
```

```python
# Generate Fibonacci series up to n terms

n = int(input("Enter number of terms: "))
a, b = 0, 1
count = 0

print("Fibonacci Series:")
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
print()
```

```python
# Check if a string is palindrome (reads same forwards and backwards)

text = input("Enter a string: ").lower()
is_palindrome = True

# Remove spaces and punctuation (optional)
clean_text = ''.join(char for char in text if char.isalnum())

for i in range(len(clean_text) // 2):
    if clean_text[i] != clean_text[len(clean_text) - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print(f"'{text}' is a palindrome")
else:
    print(f"'{text}' is not a palindrome")

```

```python
# Password must be at least 8 characters, contain a number and a special char

password = input("Enter password: ")
special_chars = "!@#$%^&*()_+-=[]{}|;:,.<>?"

if len(password) < 8:
    print("Password must be at least 8 characters long")
elif not any(char.isdigit() for char in password):
    print("Password must contain at least one number")
elif not any(char in special_chars for char in password):
    print("Password must contain at least one special character")
else:
    print("Password is valid!")
```

```python
# Implement Rock Paper Scissors game

import random

choices = ["rock", "paper", "scissors"]
computer = random.choice(choices)
user = input("Enter rock, paper, or scissors: ").lower()

if user not in choices:
    print("Invalid choice!")
else:
    print("Computer chose:", computer)

    if user == computer:
        print("It's a tie!")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win!")
    else:
        print("You lose!")
```
